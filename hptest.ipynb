```python
def cat_encoder(df,variable):
    test[variable]=le.fit_transform(test[variable])
```

```python
for feature in ['LotShape','SaleCondition','LandSlope','OverallQual','OverallCond','ExterQual','ExterCond','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','HeatingQC','KitchenQual','FireplaceQu','GarageQual','GarageCond','PoolQC','Fence']:
    cat_encoder(test, feature)
```

```python
test.head(10)
```

```python
def cat_onehot(test,variable):
    variable_new = pd.get_dummies(test[variable],drop_first=True)
    test.drop([variable], axis=1, inplace=True)
    test=pd.concat([test,variable_new],axis=1)
```

```python
for feature in ['LandContour','Utilities','LotConfig','BldgType','RoofStyle','MasVnrType','Foundation','Heating','CentralAir','Electrical','GarageType','GarageFinish','PavedDrive','MiscFeature']:
    cat_onehot(test,feature)
```
